```python
#import necessary libraries
import pandas as pd
import numpy as np
import os
```

```python
current_directory  = os.getcwd()
#extract final result data
final_result = pd.read_csv(current_directory + '/final result.csv')
final_result.head()
```

```python
lab_investigation = pd.read_csv(current_directory + '/Laboratory investigation.csv')
lab_investigation.head()
```

```python
#extracting only positive HIV tests
final_result_positive = final_result[(final_result['test'].str.lower() == 'hiv') 
                                     & (final_result['result'].str.lower() == 'positive')]\
                                    [['person_investigation_id', 'date', 'result']]\
                                    .rename(columns={'result': 'final_result', 'person_investigation_id':'person_id'})


final_result_positive.head()
```

```python
lab_investigation_results = lab_investigation[['laboratory_investigation_id', 'result', 'time']]\
                            .rename(columns={'laboratory_investigation_id': 'person_id'})

lab_investigation_results.head()
```

```python
# Group rows by 'Subject' and add a new column to store grouped results
lab_investigation_results['grouped_results'] = \
lab_investigation_results.groupby('person_id')['time'].transform(lambda x: '|'.join(x.astype(str)))

lab_investigation_results
```

```python
# Merge the DataFrames on the 'Subject' column
merged_tests = pd.merge(final_result_positive, lab_investigation_results, on='person_id', how='inner')
merged_tests.head()
```

```python
merged_tests.drop('grouped_results', axis=1, inplace=True)
```

```python
merged_tests.to_excel(current_directory+ '/testexcel.xlsx', index=False)
```

```python
# Pivot the DataFrame to show records of one ID in one row
merged_tests_pivoted_df = merged_tests.pivot_table(index='person_id', columns= 'time', values='result', aggfunc='first')

# Reset the index to make 'ID' a regular column
merged_tests_pivoted_df.reset_index(inplace=True)

merged_tests_pivoted_df.head()
```

```python
merged_tests['grouped_results'] = merged_tests['grouped_results'].str.split('|') 
merged_tests.head()
```

```python

#Determine the length of the longest list
max_len = merged_tests['grouped_results'].apply(len).max()

# Expand the lists into separate columns, filling shorter lists with NaN
merged_tests_expanded = pd.DataFrame(merged_tests['grouped_results'].apply(lambda x: x + [np.nan]*(max_len-len(x))).tolist())


#  Rename the new columns (optional)
merged_tests_expanded.columns = [f'element_{i+1}' for i in range(max_len)] 


#  Combine with the original DataFrame (optional if you need to keep other columns)
merged_tests_final = pd.concat([merged_tests, merged_tests_expanded], axis=1)


merged_tests_final.head()
```
